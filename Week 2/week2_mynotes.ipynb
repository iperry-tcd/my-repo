```python
# dictionaries
volley_player = {"name": "isabella", 
                "age": 23,
                "position": "oposite",
                "height": 169}
# access value
volley_player["name"]

# add/update key value pair
volley_player["position"] = "centre"
volley_player["position"]

# remove key value pair
del volley_player["height"]
print(volley_player)
volley_player["height"] = 169
print(volley_player)

# methods
print(volley_player.keys())
print(volley_player.values())
print(volley_player.items())
volley_player.get("age")
```

```python
# list comprehension
l = [i for i in range(11)]
print(l)

# list comporehension with a condition
y = [i for i in range(11) if i%2==0]
print(y)

#dictionary comprehension
words = ['zero','one','two','three','four','five','six','seven','eight','nine','ten']
nums = [i for i in range(11)]
d = {word:num for word, num in zip(words, nums)}
print("d =", d)
d_even = {word:num for word, num in d.items() if num%2==0}
print("d_even =", d_even)


# generator expressions
gen = (i**2 for i in range(6))
print(gen)
for j in gen:
    print(j)
```
