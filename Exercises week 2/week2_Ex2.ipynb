````python
def decimal_to_binary(n):
    if n == 0:
        return "0"
    binary_str = ""
    while n > 0:
        bit = n % 2 # remainder
        binary_str = str(bit) + binary_str
        n //= 2 # drops the decimal values
    return "0b" + binary_str

decimal = 745
binary = decimal_to_binary(decimal)
print(binary)

print(bin(decimal))
````

To read the binary number we can start from right to left and can use negative indexing and the len() fn.

````python
binary = "10111001"

def binary_to_decimal(b):
    b = str(b)
    if b == "0":
        return 0
    decimal_int = 0
    position = len(b)
    power = len(b) - 1
    for i in b:
        bit = int(b[position - 1]) * (2 ** (len(b) - position))
        decimal_int += bit
        position -= 1
        power -= 1
    return decimal_int

print(binary_to_decimal(1011101001))
````

## Exercise 2: Balmer Lines

As seen in the the excercises from Week 1, the energy levels of a hydrogen atom are the following

$$ E_n = - \frac{e^2}{8 \pi \varepsilon_0 a_0 n^2} \approx \frac{-13.6\text{eV}}{n^2} .$$

Make an empty list and append the values of the first ten energies using a for-loop.

````python
E_levels = []
for i in range(10):
    E_n = - 13.6 / (i+1) ** 2
    E_levels.append(E_n)

print(E_levels)
````

````python
Balmer = [1242 / abs(i) for i in E_levels if i not in (E_levels[0], E_levels[1])]

print(Balmer)
````

## Exercise 3: Quantum Qubit Gates
It turns out we can contain lists within lists i.e. $[4.0,[1,5]]$. Using nested lists for each row, represent the following single qubit gates X,Y, and Z and print them.

$$ X = \begin{pmatrix}
0 & 1 \\ 1&0
\end{pmatrix}, \ \ \ 
Y = \begin{pmatrix}
0 & -i \\ i&0
\end{pmatrix}, \ \ \ 
Z = \begin{pmatrix}
1 & 0 \\ 0&-1
\end{pmatrix}$$


````python
X = [[0, 1], [1, 0]]
Y = [[0, -1j], [1j, 0]]
Z = [[1, 0], [0, -1]]

print("X = ", X)
print("Y = ", Y)
print("Z = ", Z)
````

There is also the Hadamard gate.
$$ H = \begin{pmatrix}
1/\sqrt{2} & 1/\sqrt{2} \\ 1/\sqrt{2} & -1/\sqrt{2}
\end{pmatrix}$$

````python
H = [[1/(2**0.5),  1/(2**0.5)], [1/(2**0.5), -1/(2**0.5)]]
print(H)
````

It is more useful if every element is a complex type so cast all matrices into complex matrices.

````python
for i in range(len(X)):
    for j in range(len(X[i])):
        X[i][j] = complex(X[i][j])

for i in range(len(Y)):
    for j in range(len(Y[i])):
        Y[i][j] = complex(Y[i][j])

for i in range(len(Z)):
    for j in range(len(Z[i])):
        Z[i][j] = complex(Z[i][j])

for i in range(len(H)):
    for j in range(len(H[i])):
        H[i][j] = complex(H[i][j])

print("X = ", X)
print("Y = ", Y)
print("Z = ", Z)
print("H = ", H)
````

Make a copy of the $Y$ matrix. Start with a $2 \times 2$ matrix of all $0$s.

````python
Y_copy = [[0,0],[0,0]]

for i in range(len(Y)):
    for j in range(len(Y[i])):
        Y_copy[i][j] = Y[i][j]

print(Y_copy)
````

Create a smiliar cell that makes a transpose of $Y$, $Y^T$.

When transposing the diagonal elements stay the same.

````python
Y_trans = [[0,0],[0,0]]

for i in range(len(Y)):
    for j in range(len(Y[i])):
        Y_trans[i][j] = Y[j][i]

print(Y_trans)
````

Now take a Hermitian adjoint and check if $Y$ is Hermitian.

````python
import numpy as np
Y_dag = [[0,0],[0,0]]

for i in range(len(Y)):
    for j in range(len(Y[i])):
        Y_dag[i][j] = np.conj(Y_trans[i][j])

if Y == Y_dag:
    print("Y is hermitian.")
else:
    print("Y is not hermitian or I did something wrong.")
````

Now multiply $X$ and $H$ to get $XH$. 


````python
def matrix_multiplication(A, B):
    multiplication = [[A[0][0]*B[0][0] + A[0][1]*B[1][0] , A[0][0]*B[0][1] + A[0][1]*B[1][1]], 
                      [A[1][0]*B[0][0] + A[1][1]*B[1][0] , A[1][0]*B[0][1] + A[1][1]*B[1][1]]]
    
    return multiplication

XH = matrix_multiplication(X, H)
````

Now multiply $XH$ and $H$ to get $HXH$. 

````python
HXH = matrix_multiplication(XH, H)

print(HXH)
````

We have verified the well known identity $$HXH = Z.$$

When counting the days lived, we need to take into account leap days and days in the months of the year. Lets make two list with the number of days in each month for a regular year, and then a leap year.

````python
Months = [31,28,31,30,31,30,31,31,30,31,30,31]
Leap_Months = [31,29,31,30,31,30,31,31,30,31,30,31]
````

Calculate whether or not your birth year and the current year are leap years. Assign your answers in boolean form to the variables given below.

To determine whether a year is a leap year, follow these steps:

1. If the year is evenly divisible by 4, go to step 2. Otherwise, go to step 5.

2. If the year is evenly divisible by 100, go to step 3. Otherwise, go to step 4.

3. If the year is evenly divisible by 400, go to step 4. Otherwise, go to step 5.

4. The year is a leap year (it has 366 days). 

5. The year is not a leap year (it has 365 days).

Source: https://learn.microsoft.com/en-us/troubleshoot/microsoft-365-apps/excel/determine-a-leap-year

````python
def leap_year(year):
    year_leap = True
    if (int(year) % 100 == 0 and int(year) % 400 == 0):
        return year_leap
    if (int(year) % 4 == 0 and int(year) % 100 != 0):
        return year_leap
    else:
        return False

current_year_leap = leap_year(current_year)
birth_year_leap = leap_year(birth_year)
````

````python
birth_year_days = 0

for i in range(len(Months)):
    if birth_year_leap == False:
        birth_year_days += Months[i]
    if birth_year_leap == True:
        birth_year_days += Leap_Months[i]

print("birth year days = ",birth_year_days)        
````

````python
current_year_days = 0

for i in range(len(Months)):
    if current_year_leap == False:
        current_year_days += Months[i]
    if current_year_leap == True:
        current_year_days += Leap_Months[i]
        
print("current year days = ",current_year_days)
````

Now, using a loop, calculate the number of days in the years between your birth year and the current year. Print the answer. 

(There are 366 days in a leap year, and 365 in a regular year.)

````python
days_inbetween = 0

# list of the years in between (from 2003 to 2024)
years_inbetween = [i for i in range(2003, 2025)]

for i in range(len(years_inbetween)):
    if leap_year(years_inbetween[i]) == False: # (not a leap year)
        days_inbetween += 365
    else:
        days_inbetween += 366

        
print("Days inbetween = ",days_inbetween)
````

## Exercise 5: Rotating Vectors In 2-Dimensions

We can define a vector in two-dimensions with two values "x" and "y", where the vector points from the origin to the coordinates $$(x,y)$$ and has magnitude given by $$ \sqrt{x^2 + y^2} $$

The vector is written as a 2x1 matrix, as follows:

$$ \begin{pmatrix}
x \\
y
\end{pmatrix}$$

We can rotate the vector around the origin by multiplying it by a matrix, namely a rotational matrix, given by:

$$ \begin{pmatrix}
\cos\theta & -\sin\theta \\
\sin\theta & \cos\theta
\end{pmatrix}$$

If we want to rotate by an angle $ \theta = \pi/2 $ then we obtain the rotation matrix:

$$ \begin{pmatrix}
0 & -1 \\
1& 0
\end{pmatrix}$$

Find the resulting vector after rotating the vector, with $x=1$ and $y=1$, by $\pi/2$. Print the answer.

````python
Vector = [1,1]
Rotation = [[0,-1],[1,0]]

def vector_rotation(R, V):
    result = [0, 0]
    row_1 = R[0][0]*V[0] + R[0][1]*V[1]
    row_2 = R[1][0]*V[0] + R[1][1]*V[1]
    result[0] += row_1
    result[1] += row_2
    return result

Rot_Vector = vector_rotation(Rotation, Vector)


print(Rot_Vector)
````

Calculate the magnitude of the initial vector and the resulting vector. Check if they are equal, print a boolean answer.

````python
def vector_mag(V):
    mag = (V[0]**2 + V[1]**2)**(1/2)
    return mag

Vector_Magnitude = vector_mag(Vector)
Rot_Vector_Magnitude = vector_mag(Rot_Vector)

if Vector_Magnitude == Rot_Vector_Magnitude:
    print("They have the same magnitude.")
else:
    print("They do not have the same magnitude.") 
````

## Exercise 6: Quantum Qubit State Normalisation

Read in the file "qubits.txt", each line represents the state of a different qubit in a two-dimesnional basis, so there are two space-seperated integers. For example,

$$236 \text{ }\text{ } 3284 $$

However, these states are not normalised. What is normalisation? In simple mathematical terms, it means the squared-sum of these integers is equal to one. In the above case, this is not true

$$236^2 + 3284^2 = 10840352 $$

To normalise, we divide each integer, $a$ and $b$, by 
$$ \sqrt{a^2 + b^2} $$

So in the above case, we change the space-seperated integers to

$$\frac{236}{\sqrt{10840352}} \ \ \frac{3284}{\sqrt{10840352}} = 0.072 \ \ 0.997 \text{  (rounding to three decimal places)} $$

Now their squared sum is

$$\frac{236^2}{10840352} + \frac{3284^2}{10840352} = \frac{10840352}{10840352} = 1$$

and we can say they are normalised. 

Read in the file, and line by line, normalise each qubit state, write the result to the file "normalised_qubits.txt" and print the answers on the screen. Round to five decimal places.

````python
def normalise_qubits(Q):
    normalised_qubits = [0, 0]
    modulus = (Q[0]**2 + Q[1]**2)**(1/2)
    
    normalised_qubits[0] += Q[0] / modulus
    normalised_qubits[1] += Q[1] / modulus
    normalised_qubits[0] = round(normalised_qubits[0], 5)
    normalised_qubits[1] = round(normalised_qubits[1], 5)
    
    return normalised_qubits

first_column = []
second_column = []

with open('C:/Users/isabe/Desktop/Trinity Modules/Python Module/pykm\my-repo/Exercises week 2/qubits.txt', 'r') as f:
    for line in f:
        bits = line.split()
        column_1 = int(bits[0])
        column_2 = int(bits[1])
        first_column.append(column_1)
        second_column.append(column_2)

# write as qubit states
qubit_states = []

for i in range(len(first_column)):
    state = [first_column[i], second_column[i]]
    qubit_states.append(state)

# normalise qubit states
normalised_qubit_states = []

for i in range(len(qubit_states)):
    normalised_qubit_states.append(normalise_qubits(qubit_states[i]))
print(normalised_qubit_states)

# write to a file

with open('C:/Users/isabe/Desktop/Trinity Modules/Python Module/pykm\my-repo/Exercises week 2/normalised_qubits.txt', 'w') as f:
    for i in normalised_qubit_states:
        f.write(str(i[0]) + "  " + str(i[1]) + "\n")
````
