```python
import numpy as np
import scipy as sp
import matplotlib.pyplot as plt
```

## Numpy ##

```python
arr_1 = np.array([5, 2], int)
arr_2 = np.array([[5, 2], [1, 1]], float)
arr_3 = np.array([[[1+2j, 3+4j], [5+6j, 7+8j]], [[9+10j, 11+12j], [13+14j, 15+16j]]], dtype=complex)

print(arr_1.dtype)
print(arr_2.dtype)
print(arr_3.dtype)

print(arr_1.shape)
print(arr_2.shape)
print(arr_3.shape)

# indexing
print(arr_1[0])
print(arr_2[0, 1])
print(arr_3[1, 0, 1])
```

```python
print(arr_1[arr_1 > 1])
print(arr_3[arr_3 > 5])

# array creation functions
indentity_matrix = np.eye(3)
one_matrix = np.ones([2, 4])
array_01 = np.linspace(0, 1, 10)
random_array = np.random.rand(5)

print(indentity_matrix)
print(one_matrix)
print(array_01)
print(random_array)
```

```python
arr_10 = np.array([[1, 1], [5, 1]])
arr_11 = np.array([[1, 0], [5, 1]])

print(arr_10 * arr_11)
print(np.max(arr_10))

inverse_10 = np.linalg.inv(arr_10)
det_10 = np.linalg.det(arr_10)
evals_10 = np.linalg.eig(arr_10)

print(inverse_10)
print(det_10)
print(evals_10)
```

```python
horizontal_stack = np.hstack((arr_10, arr_11))
vertical_stack = np.vstack((arr_10, arr_11))
split_array = np.split(vertical_stack, 2)

print(horizontal_stack)
print(split_array)
```

```python
arr_20 = np.array([2, 6, 2, 8, 2, 2, 1, 7])
sorted_arr_20 = np.sort(arr_20)
print(sorted_arr_20)
print(arr_20.argsort())
```

```python
x_1 = np.linspace(-5, 5, 100)
y_1 = 2 * x_1 ** 2 + x_1 + np.random.normal(0, 10, size=x_1.shape)

coefficients = np.polyfit(x_1, y_1, 2)
y_fit = np.polyval(coefficients, x_1)

print(coefficients)
```

```python
x_vals = np.linspace(0, 2*np.pi, 100)
y_1 = np.sin(x_vals)
y_2 = np.cos(x_vals)

plt.plot(x_vals, y_1, color = "red", lw = "7", label = "$\sin(x)$")
plt.plot(x_vals, y_2, color = "blue", ls = ":", label = "$\cos(x)$")
plt.title("$\sin(x)$ and $\cos(x)$")
plt.grid()
plt.xlabel("x")
plt.ylabel("y")
plt.legend()

plt.show()
```

```python
# logarithmic scale
plt.loglog(x_vals, y_1)
plt.loglog(x_vals, y_2)

plt.show()
```

```python
# %matplotlib notebook
# command that can make our plots interactive

x = np.linspace(0, 8*np.pi, 100)

y_1 = np.sin(x)
y_2 = x * np.sin(x)
y_3 = np.cos(x)
y_4 = -np.sin(x**2)

plt.suptitle("Trigometric functions")

plt.subplot(221)
plt.plot(x, y_1)

plt.subplot(222)
plt.plot(x, y_2)

plt.subplot(223)
plt.plot(x, y_3)

plt.subplot(224)
plt.plot(x, y_3)

plt.show()
```
