```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.arange(-5, 5, .01)
y = np.arange(-5, 5, .01)

def oscillator_potential(x, y, m, w):
    return 0.5 * m * w**2 * (x**2 + y**2)

Xmesh, Ymesh = np.meshgrid(x, y)
Zmesh = oscillator_potential(Xmesh, Ymesh, 1, 1)

fig = plt.figure(figsize=(13, 13))
ax1 = fig.add_subplot(1, 1, 1, projection = "3d")
ax1.plot_surface(Xmesh, Ymesh, Zmesh)

plt.show()
```

```python
x = np.arange(-5, 5, .01)
y = np.arange(-5, 5, .01)

def ground_state(x, y, m, w):
    return (((m*w)/np.pi)**0.5) * np.e**(-((m * w) / 2) * (x ** 2 + y ** 2)) 

Xmesh, Ymesh = np.meshgrid(x, y)
Zmesh = np.abs(ground_state(Xmesh, Ymesh, 1, 1))**2

plt.figure(figsize=(8, 6))
heatmap = plt.pcolormesh(Xmesh, Ymesh, Zmesh, shading = "auto", cmap = "viridis")
plt.colorbar(heatmap)

plt.show()
```
