Write a function that takes a list of numbers as input and returns the sum of all the numbers in the list.

```python
def list_sum(list):
    sum = 0
    for i in list:
        sum += i
    return sum
```

Write a function that takes a string as input and returns the string in reverse order.

```python
def reverse_string(string):
    reversed_string = ""
    while len(string) > 0:
        reversed_string = reversed_string + string[-1]
        string = string[:-1]
    return reversed_string

print(reverse_string("Hello"))
```

Write a function that takes a list of strings as input and returns a new list containing only the strings that have a length greater than 3.

```python
def join_strings(strings):
    for i in strings:
        if len(i) > 3:
            return i
```

Write a recursive function that takes a positive integer n as input and returns the nth Fibonacci number. Hint: To print the Fibonacci sequence in Python, we need to generate a series of numbers where each number is the sum of the two preceding ones, starting from 0 and 1. The Fibonacci sequence follows a specific pattern that begins with 0 and 1, and every subsequent number is the sum of the two previous numbers.

$F(n) = F(n - 1) + F(n - 2)$

```python
def fibonacci(n):
    if n == 0:
        return 0
    elif n == 1:
        return 1
    else:
        return fibonacci(n - 1) + fibonacci(n - 2)
```

Write a function that takes a list of numbers as input and returns a new list containing the squares of all the numbers in the original list. Use the map() function and a lambda function to accomplish this.

```python
def squares(numbers):
    return list(map(lambda x: x**2, numbers))
    
print(squares(numbers))
```

Write a function that takes a list of tuples, where each tuple contains a name and an age, and returns a new list of names sorted by age in ascending order. Use the sorted() function and a lambda function to accomplish this.

```python
def age_ascending(tuples):
    return [name for name, age in sorted(tuples, key=lambda age: tuples[1])]
```

Write a function that takes a list of numbers as input and returns the maximum number in the list. Do not use the built-in max() function.

```python
def find_max(numbers):
    sorted_list = [i for i in sorted(numbers)]
    return sorted_list[-1]
```

Write a function that takes a list of numbers as input and returns the average of all the numbers in the list.

```python
def average(number_list):
    number_sum = 0
    for i in number_list:
        number_sum += i
    average = number_sum / len(number_list)
    return average
```

Create a class Rectangle that has attributes for width and height. Include methods to calculate the area and perimeter of the rectangle.

```python
class Rectangle:
    def __init__(self, width, height):
        self.width = width
        self.height = height

    def area(self):
        return self.width * self.height

    def perimeter(self):
        return 2 * self.width + 2 * self.height

myr = Rectangle(2, 3)
print(myr.area())
print(myr.perimeter())
```

Create a class BankAccount that has attributes for account holder’s name and balance. Include methods to deposit, withdraw, and check the balance.

```python
class BankAccount:
    def __init__(self, name, balance):
        self.name = name
        self.balance = balance

    def deposit(self, amount):
        self.balance += amount
        return self.balance

    def withdraw(self, amount):
        self.balance -= amount
        return self.balance

    def check_balance(self):
        print(self.balance)

person = BankAccount("Bella", 100)
person.check_balance()
person.deposit(200)
person.check_balance()
person.withdraw(50)
```

```python
class Particle:
    def __init__(self, mass, charge, position):
        self.mass = mass
        self.charge = charge
        self.position = position

    def kinetic(self, v):
        self.v = v
        return 0.5 * self.mass * v ** 2

    def potential(self, h):
        self.h = h
        return self.mass * 9.81 * self.h

particle = Particle(5, 2, (1, 1, 1))
print(particle.kinetic(10))
print(particle.potential(10))
```

```python
class ElectricField:
    def __init__(self, strength, direction):
        self.strength = strength
        self.direction = direction

    def Force(self, particle):
        force = [particle.charge * self.strength * i for i in self.direction]
        return force

my_particle = Particle(5, 2, (1, 2, 1))
field = ElectricField(10, (1, 2, 1))
print(field.Force(my_particle))
```

```python
import numpy as np
import scipy

class QuantumState:
    def __init__(self, wavefunction, energylevel):
        self.wavefunction = wavefunction
        self.energylevel = energylevel

    def prob_density(self, x):
        wf = self.wavefunction(x)
        return abs(wf) ** 2

    def position_exp_val(self, x_interval):
        integrand = lambda x: x * (self.wavefunction(x) ** 2)
        return scipy.integrate.quad(integrand, x_interval[0], x_interval[1])

wavefunction = lambda x: (2 / np.pi)**0.25 * np.exp(-x**2)
quantum_state = QuantumState(wavefunction, 1)
print(quantum_state.prob_density(0))
print(quantum_state.position_exp_val((0, 2)))
```

```python
class Circuit:
    def __init__(self, resistance, capacitance, inductance):
        self.resistance = resistance
        self.capacitance = capacitance
        self.inductance = inductance

    def impedance(self, frequency):
        inductive_reactance = 2 * np.pi * frequency * self.inductance
        capacitive_reactance = 1 / (2 * np.pi * frequency * self.capacitance)
        return np.sqrt(self.resistance ** 2 + (inductive_reactance - capacitive_reactance) ** 2)

    def resonant_frequency(self):
        return 1 / (2 * np.pi * np.sqrt(self.inductance * self.capacitance))

circuit = Circuit(100, 1e-6, 1e-3)
print(circuit.impedance(1000))
print(circuit.resonant_frequency())
```

```python
class QuantumSystem:
    def __init__(self, hamiltonian, wavefunction):
        self.hamiltonian = hamiltonian
        self.wavefunction = wavefunction

    def time_evolution(self, t):
        evol_operator = lambda t: np.e ** ((-1j * self.hamiltonian * t) / np.hbar)
        # assuming the wf given is at t = 0
        return evol_operator * self.wavefunction

    def expectation_value(self, observable, interval):
        integrand = lambda x: np.conjugate(self.wavefunction(x)) * observable(x) * self.wavefunction(x)
        return scipy.integrate.quad(integrand, interval[0], interval[1])

class HarmonicOscillator(QuantumSystem):
    def __init__(self, mass, frequency):
        hamiltonian = lambda x: (1/2) * mass * (frequency**2) * (x**2)
        wavefunction = lambda x: (mass * frequency / 3.14159)**0.25 * (2.71828)**(-mass * frequency * (x**2) / 2)  # Ground state wavefunction
        super().__init__(hamiltonian, wavefunction)

oscillator = HarmonicOscillator(9.11e-31, 1e14)
oscillator.expectation_value(lambda x: x, (-1e-9, 1e-9))
```

```python
import numpy as np
import scipy.integrate

class QuantumSystem:
    def __init__(self, hamiltonian, wavefunction):
        self.hamiltonian = hamiltonian
        self.wavefunction = wavefunction

    def time_evolution(self, t):
        # Simplified: treating Hamiltonian as a scalar for illustration
        evol_operator = np.exp((-1j * self.hamiltonian(0) * t) / np.hbar)
        return lambda x: evol_operator * self.wavefunction(x)

    def expectation_value(self, observable, interval):
        integrand = lambda x: np.conjugate(self.wavefunction(x)) * observable(x) * self.wavefunction(x)
        result, _ = scipy.integrate.quad(integrand, interval[0], interval[1])
        return result

class HarmonicOscillator(QuantumSystem):
    def __init__(self, mass, frequency):
        hamiltonian = lambda x: 0.5 * mass * (frequency**2) * (x**2)
        wavefunction = lambda x: (mass * frequency / np.pi)**0.25 * np.exp(-mass * frequency * (x**2) / 2)
        super().__init__(hamiltonian, wavefunction)

oscillator = HarmonicOscillator(9.11e-31, 1e14)
print(oscillator.expectation_value(lambda x: x, (-1e-9, 1e-9)))
```
